```python
import torch
import torch.nn as nn
import math
```

```python
class SingleAttention(nn.Module):
    def __init__(self):
        super().__init__()

        self.query_linear = nn.Linear(128, 64)
        self.value_linear = nn.Linear(128, 64)
        self.key_linear = nn.Linear(128, 64)

        self.out_linear = nn.Linear(64, 64)

    def forward(self, x):
        # (32, 10, 128)
        query = self.query_linear(x) # (32, 10, 64)
        key = self.key_linear(x)
        value = self.value_linear(x)

        # print(query.shape, key.transpose(-2, -1).shape)

        x = torch.matmul(query, key.transpose(-2, -1)) / math.sqrt(128)
        # print(x.shape)
        x = torch.softmax(x, dim=-1)
        x = torch.matmul(x, value)
        x = self.out_linear(x)
        return x
```

Using Pytorch nn.multithreadattention

```python
class MultiHeadedAttention(nn.Module):
    def __init__(self):
        super().__init__()

        self.input_proj = nn.Linear(128, 64)
        self.attn = nn.MultiheadAttention(embed_dim=64, num_heads=1, batch_first=True)
        self.out_proj = nn.Linear(64, 64)

    def forward(self, x):
        x_proj = self.input_proj(x)
        attn_output, _ = self.attn(x_proj, x_proj, x_proj)  # Self-attention
        out = self.out_proj(attn_output)
        return out
```

```python
embed_dim = 128
num_heads = 8
x = torch.randn(32, 10, embed_dim)
# x: [batch_size, seq_len, embed_dim]

self_attention = MultiHeadedAttention()
out = self_attention(x)
print(out.shape)  # Expected output: (32, 10, 128)
```
